```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("dataset_sdn.csv")
```

```python
data[data['label'] == 1].head()
```

```python
data[data['label'] == 0].head()
```

```python
data = data.dropna()
```

```python
columns = []
for i in data.columns:
    #print(data[i].dtype == 'O')
    if data[i].dtype != 'O':
        columns.append(i)
print(columns)
```

```python
data = data[columns]
```

```python
label = data["label"]
data = data.drop(["label"], axis=1)
```

```python
# plt.figure(figsize=(15,8))
# ax = sns.heatmap(data.corr(), annot=True);
```

```python
data.corr()[(data.corr() > 0.70) == True];
```

```python
data.corr().columns.shape;
```

```python
accepted_cols = ['pktcount', 'bytecount','tot_dur', 'packetins', 'pktperflow', 'byteperflow', 'pktrate', 'Pairflow','tx_bytes', 'rx_bytes', 'tx_kbps', 'rx_kbps', 'tot_kbps']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import BernoulliRBM
```

```python
train_X, test_X, train_y, test_y = train_test_split(data[accepted_cols], label, test_size=0.30)
```

```python
models = [LogisticRegression, SGDClassifier, MLPClassifier, BernoulliRBM]
```

```python
model = LogisticRegression(max_iter=2000)
```

```python
model.fit(train_X, train_y)
```

```python
model.score(train_X, train_y)
```

```python
model.score(test_X, test_y)
```

```python
for model in models:
    model_name = model.__name__
    model = model(max_iter=500)
    model.fit(train_X, train_y)
    print("Model Name:" + model_name)
    print("Train Score:" + str(model.score(train_X, train_y)))
    print("Test Score: " + str(model.score(test_X, test_y)))
```
